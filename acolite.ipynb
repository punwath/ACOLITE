```python
#Using Terminal to run these prompts
#cd into a suitable directory and clone the git repository: git clone --depth 1 https://github.com/acolite/acolite
#conda create -n acolite -c conda-forge python=3
#conda activate acolite
#conda install -c conda-forge numpy matplotlib scipy gdal libgdal-jp2openjpeg libgdal-netcdf pyproj scikit-image pyhdf pyresample netcdf4 h5py requests pygrib cartopy zarr fsspec aiohttp


import sys, glob, os
sys.path.append(r'C:\Users\ppp13\acolite')
import acolite as ac

input_dir  = r'C:\Users\ppp13\Downloads\Landsat images'
output_dir = r'C:\Users\ppp13\Downloads\acolite\output'

scenes = glob.glob(os.path.join(input_dir, '*.tar'))
print(f'Found {len(scenes)} scenes')

for i, scene in enumerate(scenes, 1):
    print(f'\n[{i}/{len(scenes)}] Processing {os.path.basename(scene)}')
    settings = {
        'inputfile': scene,
        'output': output_dir,
        'aerosol_correction': 'dark_spectrum',
        'dsf_aot_estimate': 'tiled',
        'l2w_parameters': ['rhos_*', 'spm_nechad2016', 'tur_nechad2016'],
    }
    ac.acolite.acolite_run(settings=settings)

print('\nDone.')
```
